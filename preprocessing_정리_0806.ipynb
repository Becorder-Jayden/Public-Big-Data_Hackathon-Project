### 업태구분 기준 (일단 임의로)

<b>--> 한식, 양식, 중식, 일식 등으로 업태구분명을 기준에 따라 재분류 필요</b>

- 양식: 경양식, 패밀리레스트랑, 패스트푸드, 제과점영업
- 한식 : 한식, 식육(숯불구이), 통닭(치킨), 탕류(보신용), 냉면집
- 중식 : 중국식
- 카페 : 까페, 전통찻집, 커피숍
- 분식 : 분식, 김밥(도시락)
- 일식 : 일식, 회집, 횟집, 복어취급
- 기타 : 기타 휴게음식점, 일반조리판매, 즉석판매제조가공업, 외국음식전문점(인도, 태국등), 뷔페식

```python
import requests
import re
import html
from bs4 import BeautifulSoup
```

```python
# 1) html 불러오는 함수 정의 

def get_html(url):
   _html = ""
   resp = requests.get(url)
   if resp.status_code == 200:
      _html = resp.text
   return _html
```

```python
# 2) 가맹점 이름 크롤링

merchant_name = []
url = 'https://www.purmeecard.com/public.do?request=merchantSelect&gu=&name=&si=08&curPage='

for i in range(1,29):
    n_url = f'{url}{i}'
    html = get_html(n_url)
    soup = BeautifulSoup(html, 'html.parser')
    b = soup.find_all('h3')
    tmp = []
    for i in b:
        tmp.append(i.string)
    merchant_name.append(tmp) 
```

```python
m_list = [name for names in merchant_name for name in names]
merchant = pd.DataFrame(m_list)
merchant.columns = ['m_name']
merchant.head()
```

```python
# 3) 주소 크롤링

merchant_address = []
url = 'https://www.purmeecard.com/public.do?request=merchantSelect&gu=&name=&si=08&curPage='

for i in range(1,29):
    n_url = f'{url}{i}'
    html = get_html(n_url)
    soup = BeautifulSoup(html, 'html.parser')
    b = soup.find_all('p',{'class' : 'text-left'})
    tmp = []
    for i in b:
        tmp.append(i.string)
    merchant_address.append(tmp)
```

```python
ma_list = [name for names in merchant_address for name in names] #리스트안의 리스트 해체
ma_list
```

```python
# 텍스트 처리
result = []
for i in ma_list:
    tmp = i.replace("\xa0\xa0","")
    result.append(tmp)
```

```python
ma = pd.DataFrame(result)
ma.columns = ['address']
len(ma)
```

```python
merchant_list = pd.concat([merchant,ma], axis = 1)
#merchant_list.to_csv('가맹점list.csv', encoding = 'cp949'
```
